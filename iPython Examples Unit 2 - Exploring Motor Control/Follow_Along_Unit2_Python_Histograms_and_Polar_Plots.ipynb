```python
#Helper code we need to get started
%matplotlib inline
import numpy as np
import matplotlib.pyplot as p
```

```python
data = np.random.rand(1000)
```

```python
p.plot(data)
```

```python
p.hist(data)
```

```python
# randn for normal distribution
datan= np.random.randn(1000)
p.hist(datan)
```

```python
help(p.hist)
```

```python
p.hist(datan, 20,cumulative=True)
```

```python
p.hist(datan, 20,normed=True,cumulative=True,color='r')
p.xlabel('Random Normalized Data')
p.ylabel('Number of points')
p.title('Probability Density Function of the Data')
```

```python
x = np.arange(0,360,45)
x
```

```python
spikecounts = np.random.rand(8)*20
```

```python
p.bar(x,spikecounts)
```

```python
help(p.bar)
```

```python
p.bar(x,spikecounts,width=45)
```

```python
p.bar(x,spikecounts,width=45)
p.xlabel('Direction')
p.ylabel('Spike Counts')
p.title('Tuning Curve')
p.axis([0,360,0,20])
```

```python
p.polar(spikecounts)
```

```python
help(p.polar)
```

```python
# fix radians issue
p.polar(x*np.pi/180, spikecounts)
```

```python
# not connecting 315 to 360 
# so append data set
spikecounts2 = np.append(spikecounts,spikecounts[0])
# convert "x-axis" / theta (independent variable) to radians
theta = np.arange(0,361,45)*np.pi/180
p.polar(theta, spikecounts2)
p.title('Tuning Curve')
```
