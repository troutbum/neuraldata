#Power Spectra and Spectrograms

In this unit, we would like you to create **power spectra** and **spectrograms** to display the data.

By the end of this video, you should be able to:

* create your own power spectra

* create your own spectrogram

* manipulate parameters to modify the power spectra and spectrograms

```python
#Helper code we need to get started
%matplotlib inline
import numpy as np
import matplotlib.pyplot as p
import wave as wave
```

```python
t = np.arange(0,1,0.001)    # 0 to 1 in millesecond steps
x = np.sin(2*np.pi*100*t)   # sin wave 100 Hz
p.plot(t,x)
p.xlabel('Time')
p.ylabel('Amplitude')
```

```python
help(p.psd)  # power spectral density to analyze frequencies
```

```python
Pxx, freqs = p.psd(x,Fs=1000)  # sample freq = 1000 Hz
```

```python
Pxx, freqs = p.psd(x,NFFT=512,Fs=1000)  # tighter versus default NFFT=256
```

```python
help(p.specgram)
```

```python
Pxx, freqs, bins, im = p.specgram(x,Fs=1000) 
p.xlabel('Time')
p.ylabel('Frequency')
```

```python
Pxx, freqs, bins, im = p.specgram(x,NFFT=1024,Fs=1000) 
p.xlabel('Time')
p.ylabel('Frequency')
```

```python
wf = wave.open('TestFile.wav') # single channel audio file
signal = wf.readframes(-1)  # convert
signal = np.fromstring(signal,'Int16') # string to numbers
p.plot(signal)
p.xlabel('Samples')
p.ylabel('Amplitude')
```

```python
fs = wf.getframerate() # sampling frequency is 44.1KHz (8 second recording)
print fs
times = np.linspace(0,len(signal)/fs,num=len(signal)) #similar to arange
p.plot(times,signal)
```

```python
Pxx, freqs = p.psd(signal, NFFT=4096, Fs=fs)  # frequency analysis
```

```python
Pxx, freqs, bins, im = p.specgram(signal,NFFT=512,Fs=fs)  #visualize frequency changes over time
```
